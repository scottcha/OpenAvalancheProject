Differences between UAC and NWAC Data
NWAC only does one dayly danger per elevation and no rose info, UAC does a full rose
    #unify by using NWAC directly and picking higest avy value per elevation for UAC
NWAC only provides a binary value on the rose for avy problems, UAC does a full rose
    #unify by using NWAC directly and picking True for the rose for values Considerable or Higher from UAC
NWAC has end dates for warning/watches, UAC sometimes does but its hard to parse out
    #Join the data if it exists--we'll likely ignroe end date for watches/warnings
NWAC does two day forecasts UAC only does one

```python
#pick higest avy value per elevation for UAC
columns_to_check = ['Day1Danger_OctagonAboveTreeline*','Day1Danger_OctagonNearTreeline*','Day1Danger_OctagonBelowTreeline*']
column_to_set = ['Day1DangerAboveTreeline', 'Day1DangerNearTreeline', 'Day1DangerBelowTreeline']
danger_values = ['Low', 'Moderate', 'Considerable', 'High', 'Extreme'] #order matters, start with lowest and overwrite with higher values as they exist
for c in range(0, len(columns_to_check)):
    for d in range(0, len(danger_values)):
        tmp = uac_df.filter(regex=(columns_to_check[c])).isin([danger_values[d]])
        tmp2 = tmp[tmp.any(axis=1)]
        uac_df.loc[uac_df.index.isin(tmp2.index),column_to_set[c]] = danger_values[d]
```

```python
danger_to_binary_rose_values = ['Considerable', 'High', 'Extreme']
def map_rose_to_binary(x):
    if(x == 'no-data'):
        return x
    elif(x in danger_to_binary_rose_values):
        return 1
    else:
        return 0
```

```python
#pick True for rose Considerable or Higher
rose_columns = ['Cornices_Octagon*', 'DeepPersistentSlab_Octagon*', 'Glide_Octagon*', 'LooseDry_Octagon*', 'LooseWet_Octagon*', 'PersistantSlab_Octagon*', 'StormSlab_Octagon*', 'WetSlabs_Octagon*', 'WindSlab_Octagon*']
for c in range(0, len(rose_columns)):
    tmp = uac_df.filter(regex=(rose_columns[c]))
    uac_df.loc[:,tmp.columns] = uac_df.filter(regex=(rose_columns[c])).applymap(map_rose_to_binary)
```
