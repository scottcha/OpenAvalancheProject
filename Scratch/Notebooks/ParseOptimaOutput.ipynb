```python
#Trial 157 finished with value: 0.5557529126628227 and parameters: {'model': 'InceptionTimePlus', 'itp_num_filters': 96, 'itp_depth': 6, 'itp_dropout_rate': 0.0, 'itp_ks': 40, 'loss_func': 'None', 'learning_rate': 1e-05, 'learner_callback': 'None', 'optimizer': 'AdamWD'}. Best is trial 3 with value: 0.6508572269711261.  {'model': 'LSTM', 'hidden_size': 250, 'lstm_n_layers': 7, 'rnn_dropout': 0.8, 'bidirectional': False, 'lstm_fc_dropout': 0.6000000000000001, 'lstm_cell_type': 'LSTM', 'loss_func': 'LabelSmoothingCrossEntropyFlat', 'learning_rate': 0.0001, 'learner_callback': 'IntraClassCutMix1d', 'optimizer': 'AdamWD'}
#Trial 158 finished with value: 0.4960776990967197 and parameters: {'model': 'LSTM', 'hidden_size': 250, 'lstm_n_layers': 7, 'rnn_dropout': 0.8, 'bidirectional': False, 'lstm_fc_dropout': 0.6000000000000001, 'lstm_cell_type': 'LSTM', 'loss_func': 'LabelSmoothingCrossEntropyFlat', 'learning_rate': 0.0001, 'learner_callback': 'IntraClassCutMix1d', 'optimizer': 'AdamWD'}. Best is trial 3 with value: 0.6508572269711261. {'model': 'InceptionTimePlus', 'itp_num_filters': 64, 'itp_depth': 6, 'itp_dropout_rate': 0.2, 'itp_ks': 60, 'loss_func': 'LabelSmoothingCrossEntropyFlat', 'learning_rate': 1e-05, 'learner_callback': 'CutMix1d', 'optimizer': 'AdamNoWD'}
#Trial 35 finished with value: 0.33674572027819655 and parameters: {'model': 'TCN', 'TCN_fc_dropout': 0.0, 'TCN_conv_dropout': 0.2, 'loss_func': 'LabelSmoothingCrossEntropy', 'learning_rate': 0.0001, 'learner_callback': 'CutMix1d', 'optimizer': 'AdamNoWD'}. Best is trial 3 with value: 0.6508572269711261.  {'model': 'TCN', 'TCN_fc_dropout': 0.0, 'TCN_conv_dropout': 0.2, 'loss_func': 'LabelSmoothingCrossEntropyFlat', 'learning_rate': 0.01, 'learner_callback': 'None', 'optimizer': 'AdamWD'}

def string_to_df(input_string):
    # Initialize an empty list to store the rows of the dataframe
    df = pd.DataFrame()
    # Split the input string by newline character
    lines = input_string.split("\n")
    # Loop through each line
    for line in lines:
        #if a line contains the word "finished", then it contains the value and parameters
        if "finished" in line:
            #extract the Trial number
            trial = int(line.split("Trial ")[1].split(" finished")[0])
            #extract the value
            value = float(line.split("value: ")[1].split(" and")[0])
            #extract the param string as everything between { and }  
            param_string = "{" + line.split("{")[1].split("}")[0] + "}"
            #create a dataframe using the param string which is a dictionary in string format
            param_dict = eval(param_string)
            param_dict["value"] = value
            param_dict["trial"] = trial
            #set the trial number as the index
            param_df = pd.DataFrame(param_dict, index=[trial])
            #condatenate the value and parameters into a dataframe
            df = pd.concat([df, param_df], axis=0)
    return df

            
```
