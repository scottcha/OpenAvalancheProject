```python
from tsfresh.convenience.bindings import dask_feature_extraction_on_chunk
from tsfresh.feature_extraction.settings import ComprehensiveFCParameters
```

```python
relevant_features =[]
stride = 1  
#break this in to batches of variables to help is scale better in memory
unique_id = df['id'].unique().compute()
```

```python
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python

for i in range(0,len(unique_id),stride):
    print('On i: ' + str(i) + ' of ' + str(len(unique_id)))
    #not every variable has a sample data so map labels to the data which does exist for that variable
    subset = df[df['id'].isin(unique_id[i:i+stride])]
    #return subset, unique_kind
    
    
    features = dask_feature_extraction_on_chunk(subset.groupby(['id', 'kind']), 
                                     column_id='id',
                                     column_kind='kind',
                                     column_sort='time',
                                     column_value='value',
                                     default_fc_parameters=ComprehensiveFCParameters())
    features = features.categorize(columns=["variable"])
    features = features.reset_index(drop=True)
    feature_table = features.pivot_table(index="id", columns="variable", values="value", aggfunc="sum")
    relevant_features.append(feature_table)
    
    #relevant_features.append( extract_relevant_features(subset, y[y_column].iloc[subset.id.unique()], column_id='id', column_sort="time", column_kind="kind", column_value="value", n_jobs=n_jobs))

relevant_features_df = dd.concat(relevant_features, axis=1)
```
